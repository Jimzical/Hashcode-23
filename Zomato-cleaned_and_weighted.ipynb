```python
import pandas as pd
df = pd.read_csv("CleanZomato.csv")
df.head()
```

```python
df.columns[9:]
```

```python
dinner_ratings = df['Dinner Ratings'].to_list()
for i in range(len(dinner_ratings)):
    if dinner_ratings[i] == '-' :
        dinner_ratings[i] = 2.5
    dinner_ratings[i] = float(dinner_ratings[i])
```

```python
df['Dinner Ratings'] = pd.Series(dinner_ratings)
df['Dinner Ratings']
```

```python
dinner_reviews = df['Dinner Reviews'].to_list()
for i in range(len(dinner_reviews)):
    if dinner_reviews[i] == 0 :
        dinner_reviews[i] = 1

df['Dinner Reviews'] = pd.Series(dinner_reviews)
df['Dinner Reviews']
```

```python
delivery_ratings = df['Delivery Ratings'].to_list()
for i in range(len(delivery_ratings)):
    if delivery_ratings[i] == '-' :
        delivery_ratings[i] = 2.5
    delivery_ratings[i] = float(delivery_ratings[i])

df['Delivery Ratings'] = pd.Series(delivery_ratings)
df['Delivery Ratings'].unique()
```

```python
delivery_ratings = df['Delivery Reviews'].to_list()
for i in range(len(delivery_ratings)):
    if delivery_ratings[i] == 0 :
        delivery_ratings[i] = 1

df['Delivery Reviews'] = pd.Series(delivery_ratings)
df['Delivery Reviews']
```

```python
df
```

```python
R=df['Dinner Ratings']
v=df['Dinner Reviews']
C=df['Dinner Ratings'].mean()
m=df['Dinner Reviews'].quantile(0.70)

df['weighted_dine'] = ((R*v)+ (C*m))/(v+m)

df['weighted_dine']
```

```python
R=df['Delivery Ratings']
v=df['Delivery Reviews']
C=df['Delivery Ratings'].mean()
m=df['Delivery Reviews'].quantile(0.70)

df['weighted_deli'] = ((R*v)+ (C*m))/(v+m)

df['weighted_deli']
```

```python
#Save the the dataset to a new file
df.to_csv('CleanZomato2.csv', index=False)
```
